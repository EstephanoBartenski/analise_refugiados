```python
# transformando coluna mês/ano em datetime object
df['Mês'] = (pd.to_datetime(df['Mês/Ano'])).dt.month
df['Ano'] = (pd.to_datetime(df['Mês/Ano'])).dt.year
```

```python
df = df.drop('Mês/Ano', axis=1)
```

```python
df.sample(10)
```

```python
df.groupby('Nacionalidade')['Quantidade'].sum().sum()
```

```python
df.groupby('Nacionalidade')['Quantidade'].sum().sort_values(ascending=False)
```

```python
df = df.drop('Total de Solicitações', axis=1)
```

```python
df.head()
```

```python
ano_2019 = df.groupby(['Ano', 'Nacionalidade'])['Quantidade'].sum().to_frame().reset_index()
ano_2019 = ano_2019.loc[ano_2019['Quantidade'] > 500]
ultimos_anos = ano_2019.loc[ano_2019['Ano'] >= 2016]
ultimos_anos
```

```python
import altair as alt
```

```python
alt.Chart(ultimos_anos).mark_bar().encode(
    x=alt.X('Ano:N'),
    y=alt.Y('Quantidade:Q', stack='normalize', title='Porcentagem de solicitações', axis=alt.Axis(format='%')),
    color=alt.Color('Nacionalidade', scale=alt.Scale(scheme='set3'))
)
```

```python
estados = df.groupby(['UF'])['Quantidade'].sum().to_frame().reset_index()
estados = estados.loc[estados['Quantidade'] > 500]
estados
```

```python
alt.Chart(data=estados, title='Solicitações por UF em toda a série histórica').mark_bar().encode(
    x=alt.X('UF', title='Unidade Federativa'),
    y=alt.Y('Quantidade', title='Quantidade de solicitações'),
    color=alt.Color('UF', scale=alt.Scale(scheme='set1'))

)
```

```python
# solicitações ao longo dos anos
```

```python
anoss = df.groupby(['Ano'])['Quantidade'].sum().to_frame().reset_index()
anoss = anoss.loc[anoss['Ano'] >= 2006]
anoss
```

```python
alt.Chart(data=anoss, title='Evolução das solicitações ao longo do tempo').mark_line().encode(
    x=alt.X('Ano:N', title='Ano'),
    y=alt.Y('Quantidade:Q', title='Quantidade de solicitações'),
)
```

```python
ven = df[df['Nacionalidade'] == 'VENEZUELA'].groupby(['Nacionalidade', 'Ano'])['Quantidade'].sum().to_frame().reset_index()
ven = ven.loc[ven['Ano'] >= 2013]
ven
```

```python
alt.Chart(data=ven, title='Solicitações de venezuelanos nos últimos anos').mark_line(color='#eb4034', size=2.6).encode(
    x=alt.X('Ano:N', title='Ano', axis=alt.Axis(grid=True)),
    y=alt.Y('Quantidade:Q', title='Quantidade de solicitações')
).properties(
    width=450,
    height=350
)
```
